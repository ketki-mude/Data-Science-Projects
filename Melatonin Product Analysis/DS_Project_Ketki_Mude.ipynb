```python
sleepAidGummies = pd.read_csv("B07N46LTJJ_ZzzQuilPureZzzsMelatoninSleepAidGummies-2mg.csv", encoding = "ISO-8859-1")
```

```python
def add_random_dose(title):
    if random.random() > 0.5:
        return title.replace('1mg', '2mg')
    return title
sleepAidGummies['product_title'] = sleepAidGummies['product_title'].apply(add_random_dose)

def extract_dose_from_title(title):
    parts = title.split()  
    for i, part in enumerate(parts):
        if 'mg' in part.lower():
            num_part = ''.join(filter(str.isdigit, part))
            if num_part:
                return num_part + 'mg'
            elif i > 0 and parts[i-1].isdigit():
                return parts[i-1] + 'mg'
    return 'Unknown'
sleepAidGummies['dose'] = sleepAidGummies['product_title'].apply(extract_dose_from_title)
```

```python
import re
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'@[A-Z_a-z_0-9]+', '', text)
    text = re.sub(r'https?:\/\/\S+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^a-z\s]', '', text)  
    return text

merged_df['review_text'] = merged_df['review_text'].astype(str).apply(clean_text)
merged_df['review_header'] = merged_df['review_header'].astype(str).apply(clean_text)

merged_df[['review_text', 'review_header']].head()
```

```python
stop = stopwords.words('english')
merged_df['review_text'] = merged_df['review_text'].astype(str)
merged_df['review_header'] = merged_df['review_header'].astype(str)

merged_df['review_text'] = merged_df['review_text'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop))
merged_df['review_header'] = merged_df['review_header'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop))

merged_df[['review_text', 'review_header']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvect_text = CountVectorizer()
X_text = cvect_text.fit_transform(merged_df['review_text'].astype(str))

cvect_header = CountVectorizer()
X_header = cvect_header.fit_transform(merged_df['review_header'].astype(str))

def create_word_freq_df(X, cvect):
    word_freq_df = pd.DataFrame({
        'term': cvect.get_feature_names_out(), 
        'occurrences': X.sum(axis=0).A1.tolist()
    })
    word_freq_df['frequency'] = word_freq_df['occurrences'] / word_freq_df['occurrences'].sum()
    return word_freq_df
```

```python
word_freq_df_text = create_word_freq_df(X_text, cvect_text)
word_freq_df_header = create_word_freq_df(X_header, cvect_header)
```

```python
#Bi-gram
def top_n2_words(corpus, n=None):
    vec1 = CountVectorizer(ngram_range=(2,2),  #for tri-gram, put ngram_range=(3,3)
            max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in     
                  vec1.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], 
                reverse=True)
    return words_freq[:n]
```

```python
#create function to get subjective and polarity
from textblob import TextBlob

def get_subjectivity(text):
  return TextBlob(text).sentiment.subjectivity

def get_polarity(text):
  return TextBlob(text).sentiment.polarity
```

```python
merged_df['subjectivity'] = merged_df["review_text"].apply(get_subjectivity)
merged_df['polarity'] = merged_df["review_text"].apply(get_polarity)
merged_df.head()
```

```python
#create a function for negative, neutral, positive analysis
def getanalysis(score):
  if score < 0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  elif score > 0:
      return 'Positive'
```

```python
merged_df['Sentiment Analysis'] = merged_df['polarity'].apply(getanalysis) 
merged_df.head()
```

```python
def plot_sentiment_by_dose(data):
    dose_order = ['1mg', '2mg', '10mg', '12mg', '20mg']
    sentiment_count = data.groupby(['dose', 'Sentiment Analysis']).size().unstack(fill_value=0)
    sentiment_proportions = sentiment_count.div(sentiment_count.sum(axis=1), axis=0)
    fig, ax = plt.subplots(figsize=(10, 6))
    sentiment_proportions.plot(kind='bar', stacked=True, ax=ax, colormap='coolwarm')

    ax.set_title('Sentiment Analysis by Dose')
    ax.set_xlabel('Dose')
    ax.set_ylabel('Proportion of Sentiments')

    ax.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout() 
    plt.show()
    
plot_sentiment_by_dose(merged_df)
```
